```python
#index is end of month change to beg of month
USRealGDP = pd.read_csv (r'C:\Users\admin\OneDrive - nyu.edu\Spring 2022\Capstone\Data\USRealGDP.csv')
USRealGDP = USRealGDP.set_index(pd.to_datetime(USRealGDP['Date']))
# USRealGDP= USRealGDP.set_index(USRealGDP.index.shift(1, freq='D'))
USRealGDP['Value'] = USRealGDP['Value'].apply(lambda x: x.rstrip('T'))
USRealGDP.drop("Date",1,inplace=True)
USRealGDP.head()
```

```python
Master_Monthly = prime[['MPRIME']]
Master_Monthly['MPRIMEpct'] = prime['MPRIME'].pct_change()
Master_Monthly['MPRIMEstd'] = prime['MPRIME'].rolling(10).std()
# Master_Monthly['VIX'] = VIX['Adj Close']
# Master_Monthly['VIXpct'] = VIX['Adj Close'].pct_change()
# Master_Monthly['VIXstd'] = VIX['Adj Close'].rolling(10).std(-10)
Master_Monthly['RealGDP'] = USRealGDP['Value'].astype(float)
Master_Monthly['RealGDPpct'] = USRealGDP['Value'].astype(float).pct_change(-1)
Master_Monthly['RealGDPstd'] = USRealGDP['Value'].astype(float).rolling(10).std().shift(-10)
Master_Monthly['UNRATEpct'] = unrate['UNRATE']
Master_Monthly['UNRATEstd'] = unrate['UNRATE'].rolling(10).std(-10)
Master_Monthly['income'] = income['A229RX0']
Master_Monthly['incomepct'] = income['A229RX0'].pct_change()
Master_Monthly['incomestd'] = income['A229RX0'].rolling(10).std(-10)
# Master_Monthly['CPIpct'] = CPI['CPALTT01USM657N']
# Master_Monthly['CPIstd'] = CPI['CPALTT01USM657N'].rolling(10).std(-10)
```

```python
#Come with change percentage
bond10 = pd.read_csv (r'C:\Users\admin\OneDrive - nyu.edu\Spring 2022\Capstone\Data\United States 10-Year Bond Yield Monthly.csv')
bond10 = bond10.set_index(bond10['Date'])
bond5 = pd.read_csv (r'C:\Users\admin\OneDrive - nyu.edu\Spring 2022\Capstone\Data\United States 5-Year Bond Yield Monthly.csv')
bond5 = bond5.set_index(bond5['Date'])
bond3 = pd.read_csv (r'C:\Users\admin\OneDrive - nyu.edu\Spring 2022\Capstone\Data\United States 3-Month Bond Yield Monthly.csv')
bond3 = bond3.set_index(bond3['Date'])
```

```python
bond10['bond10'] = bond10['Price']
bond10['bond10pct'] = bond10['Change %'].apply(lambda x: x.rstrip('%')).astype(float)/100
bond10['bond10std'] = bond10['Price'].rolling(10).std().shift(-10)
bond10['bond5'] = bond5['Price']
bond10['bond5pct'] = bond5['Change %'].apply(lambda x: x.rstrip('%')).astype(float)/100
bond10['bond5std'] = bond5['Price'].rolling(10).std().shift(-10)
bond10['bond3'] = bond3['Price']
bond10['bond3pct'] = bond3['Change %'].apply(lambda x: x.rstrip('%')).astype(float)/100
bond10['bond3std'] = bond3['Price'].rolling(10).std().shift(-10)
```

```python
bond10['year'] = bond10['Date'].str[-2:].astype(int)
bond10['newdate'] = np.where( bond10['year'] >23, bond10['Date'].str[:4] + "19" + bond10['year'].astype(str),bond10['Date'])
bond10 = bond10.set_index(pd.to_datetime("01 " + bond10['newdate']))
bond10.drop(bond10.iloc[:, 0:6],1,inplace = True)
```
